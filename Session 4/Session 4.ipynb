# Matplotlib

Matplotlib is a popular Python library used for creating visualizations such as graphs, charts, and plots. It provides a wide range of customization options and supports multiple data visualization types.

Matplotlib.pyplot is a sub-library of Matplotlib that provides a MATLAB-like interface for creating visualizations. It is used for creating simple plots, histograms, scatterplots, and more. It is easy to use and provides a high-level interface for creating visualizations.

To use Matplotlib.pyplot in Python, you need to first import it using the following command:

## Figures and Subplots
Basically, the plt is a common alias of matplotlib.pyplot used by most people. When we plot something using plt such as plt.line(...), we implicitly created a Figure instance and an Axes inside the Figure object. This is totally fine and very convenient when we just want to draw a single graph.

However, we can explicitly call plt.subplots() to get the Figure object and Axes object, in order to do more things on them. When we want to draw multiple subplots on a Figure, it is usually required to use this approach.

Also, here are the Matplotlib official API reference for the Figure and Axes classes. It is highly recommended to check them out and try some methods yourselves to make sure you understand even deeper.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 2)
ax3 = fig.add_subplot(2, 3, 3)
ax4 = fig.add_subplot(2, 3, 4)
ax1.plot(x, x)
ax2.plot(x, x**3)
ax3.plot(x, np.e**x)
ax4.plot(x, x**2)
```

```python
# ADJUSTING PLOT:LINE COLOR
plt.plot(x,np.sin(x),color='blue')         # specify color by name
plt.plot(x,np.sin(x-1),color='g')          # short color code (rgbcymk)
plt.plot(x,np.sin(x-2),color='0.75')       # grayscale between 0 and 1
plt.plot(x,np.sin(x-3),color='#FFDD44')    # hex code
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3))# RGB tuple,values between 0 and 1
plt.plot(x,np.sin(x-5),color='chartreuse') # all HTML color names supported
```

```python
m_fig = plt.figure(figsize=(15, 10), facecolor = 'white')
m_axes = m_fig.gca()
m_axes.grid(alpha= 0.2)
```

```python
mataria = pd.read_csv("final_mataria.csv").fillna(0).astype(np.int16).drop("Unnamed: 0",axis=1).set_index("seat")
```

```python
bins = [0,60,70,80,90,100]
labels = ['f','d','c','b','a']
mataria.columns
```

```python
grades = pd.DataFrame()
for subject in mataria.columns:
  grades[subject] = pd.cut(mataria[subject], 
       bins = bins,
       labels = labels,
       right = True,
       include_lowest = True)
```

```python
grades.melt()
```

```python
plotting_data = grades.melt(var_name = "subject",value_name='grade' )
```
